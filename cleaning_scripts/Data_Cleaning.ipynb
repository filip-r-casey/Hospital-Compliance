```python
df_features=pd.read_csv('hospitals.csv')
df_links = pd.read_csv('machine_readable_links.csv')
```

```python
df_features.head()
```

```python
subset_features = df_features.loc[:, ['bed_count', 'zip_code','ccn','medicare_medicaid_eligible']]
subset_features.head()
print(len(subset_features))
subset_features = subset_features.dropna()
print(len(subset_features))

```

```python
df_links.head()
```

```python
subset_links = df_links.loc[:, ['ccn', 'state_or_region', 'reporting_entity_name_common', 'machine_readable_url']]
print("Before: ", len(subset_links))
subset_links = subset_links.dropna()
print("After:", len(subset_links))
```

```python
conn = psycopg2.connect(database="postgres", user="postgres", password="postgres", host="localhost", port=5432)
subset_features.to_sql('features', conn, if_exists='replace', index=False)
conn.close()
```
